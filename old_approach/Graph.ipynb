```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data0 = pd.read_csv('extracted_features0.csv')
data1 = pd.read_csv('extracted_features1.csv')
```

```python
Q1 = data0.quantile(0.25)
Q3 = data0.quantile(0.75)
IQR = Q3 - Q1
all_data = data0[~((data0 < (Q1 - 1.5 * IQR)) |(data0 > (Q3 + 1.5 * IQR))).any(axis=1)]
all_data.shape
```
